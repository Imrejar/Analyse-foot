# Premier league Data

```python
# Import des bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt

# Lecture du fichier CSV
nom_fichier = "england-premier-league-teams-2018-to-2019-stats.csv"  # Assurez-vous de spécifier le chemin correct vers votre fichier CSV
data = pd.read_csv(nom_fichier)

# Affichage des premières lignes du DataFrame pour vérifier que les données ont été correctement chargées
data.head()
```

## Classement

```python

# Tri des équipes par position dans la ligue
def sort_teams_by_league_position(teams):
    return teams.sort_values(by='league_position')

# Affichage du nom de l'équipe et de sa position dans la ligue
def display_team_name_and_position(teams):
    print(teams[['league_position','team_name', 'total_points']])

# Tri des équipes par position dans la ligue
sorted_teams = sort_teams_by_league_position(data)

# Affichage du nom de l'équipe et de sa position dans la ligue
display_team_name_and_position(sorted_teams)


```

## Tableau barre: Nombre de matches gagnés par équipe & Différence de buts par équipe

```python
# Calcul de la différence de buts
data['DifferenceButs'] = data['goals_scored'] - data['goals_conceded']

# Crée un diagramme à barres pour comparer le nombre de matches gagnés par équipe
plt.figure(figsize=(12, 8))
plt.bar(data['team_name'], data['wins'], color='skyblue')
plt.xlabel('Equipe')
plt.ylabel('Nombre de matches gagnés')
plt.title('Nombre de matches gagnés par équipe')
plt.xticks(rotation=80)
plt.show()

# Crée un diagramme à barres pour comparer la différence de buts par équipe
plt.figure(figsize=(10, 6))
plt.bar(data['team_name'], data['goal_difference'], color='lightgreen')
plt.xlabel('Equipe')
plt.ylabel('Différence de buts')
plt.title('Différence de buts par équipe')
plt.xticks(rotation=80)
plt.show()
```

## Graphe: wins/draws/losses away or home

```python
# Sélection des colonnes pertinentes pour les performances à domicile et à l'extérieur
home_away_data = data[['team_name', 'wins_home', 'draws_home', 'losses_home', 'wins_away', 'draws_away', 'losses_away']]


# Configuration des données pour les graphiques
teams = home_away_data['team_name']
wins_home = home_away_data['wins_home']
draws_home = home_away_data['draws_home']
losses_home = home_away_data['losses_home']
wins_away = home_away_data['wins_away']
draws_away = home_away_data['draws_away']
losses_away = home_away_data['losses_away']

# Création des graphiques en barres empilées
fig, axs = plt.subplots(nrows=len(teams), ncols=1, figsize=(10, 5 * len(teams)))

for i in range(len(teams)):
    axs[i].bar(['Wins', 'Draws', 'Losses'], [wins_home[i], draws_home[i], losses_home[i]], label='Home', color='blue')
    axs[i].bar(['Wins', 'Draws', 'Losses'], [wins_away[i], draws_away[i], losses_away[i]], bottom=[wins_home[i], draws_home[i], losses_home[i]], label='Away', color='orange')
    axs[i].set_title(teams[i])
    axs[i].set_ylabel('Matches')
    axs[i].legend()

plt.tight_layout()
plt.savefig('wins_draws_losses.png')

# Affichage des graphiques
plt.show()
```

## Graph2

```python
# Sélection des colonnes pertinentes pour les buts marqués et concédés

# Configuration des données pour les graphiques
teams = data['team_name']
goals_scored = data['goals_scored']
goals_conceded = data['goals_conceded']
goals_scored_home = data['goals_scored_home']
goals_conceded_home = data['goals_conceded_home']
goals_scored_away = data['goals_scored_away']
goals_conceded_away = data['goals_conceded_away']

# Création des graphiques
fig, axs = plt.subplots(nrows=len(teams), ncols=2, figsize=(15, 5 * len(teams)))

for i in range(len(teams)):
    # Graphique pour les buts marqués et concédés
    axs[i, 0].bar(['Goals Scored', 'Goals Conceded'], [goals_scored[i], goals_conceded[i]], color=['green', 'red'])
    axs[i, 0].set_title(teams[i] + ' - Total Goals')
    axs[i, 0].set_ylabel('Goals')

    # Graphique pour la répartition des buts à domicile et à l'extérieur
    axs[i, 1].bar(['Home', 'Away'], [goals_scored_home[i], goals_scored_away[i]], label='Goals Scored', color='green')
    axs[i, 1].bar(['Home', 'Away'], [goals_conceded_home[i], goals_conceded_away[i]], label='Goals Conceded', bottom=[goals_scored_home[i], goals_scored_away[i]], color='red')
    axs[i, 1].set_title(teams[i] + ' - Home vs Away Goals')
    axs[i, 1].set_ylabel('Goals')
    axs[i, 1].legend()

plt.tight_layout()
plt.savefig('goalsscoredconceded_home_away.png')

# Affichage des graphiques
plt.show()
```

## Graph goals scored conceded / quarter

```python
# Sélectionner les colonnes pertinentes
match_data_scored = data[["team_name", "goals_scored_min_0_to_10", "goals_scored_min_11_to_20", "goals_scored_min_21_to_30",
                   "goals_scored_min_31_to_40", "goals_scored_min_41_to_50", "goals_scored_min_51_to_60",
                   "goals_scored_min_61_to_70", "goals_scored_min_71_to_80", "goals_scored_min_81_to_90",
                   ]]

match_data_conceded = data[["team_name","goals_conceded_min_0_to_10", "goals_conceded_min_11_to_20", "goals_conceded_min_21_to_30",
                   "goals_conceded_min_31_to_40", "goals_conceded_min_41_to_50", "goals_conceded_min_51_to_60",
                   "goals_conceded_min_61_to_70", "goals_conceded_min_71_to_80", "goals_conceded_min_81_to_90"]]

# Définir les équipes comme index
match_data_scored.set_index("team_name", inplace=True)
match_data_conceded.set_index("team_name", inplace=True)


# Tracer un graphique à barres pour comparer les buts marqués par quart de match pour différentes équipes
match_data_scored.plot(kind="bar", figsize=(10, 10))
plt.title("Comparaison des buts marqués par quart de match pour chaque équipe")
plt.xlabel("Équipe")
plt.ylabel("Nombre de buts")
plt.xticks(rotation=80)
plt.legend(title="Quart du match")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Tracer un graphique à barres pour comparer les buts concédés par quart de match pour différentes équipes
match_data_conceded.plot(kind="bar", figsize=(10, 10))
plt.title("Comparaison des buts concédés par quart de match pour chaque équipe")
plt.xlabel("Équipe")
plt.ylabel("Nombre de buts")
plt.xticks(rotation=80)
plt.legend(title="Quart du match")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Matrice de corrélation entre stats et relation tir/but

```python
import seaborn as sns
```

```python
# Sélectionner les colonnes pertinentes pour l'analyse de corrélation
corr_data = data[["goals_scored", "shots", "corners_per_match", "fouls", "cards_per_match"]]

# Calculer la matrice de corrélation
correlation_matrix = corr_data.corr()

# Afficher la matrice de corrélation avec seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matrice de corrélation entre différentes statistiques")
plt.show()

# Utiliser des graphiques de dispersion pour visualiser les relations entre les tirs et les buts marqués
plt.figure(figsize=(8, 6))
sns.scatterplot(x="shots", y="goals_scored", data=data)
plt.title("Relation entre les tirs et les buts marqués")
plt.xlabel("Tirs")
plt.ylabel("Buts marqués")
plt.grid(True)
plt.show()
```

## Comparaison de possession  

```python
# Sélectionner les colonnes pertinentes pour l'analyse de possession de balle
possession_data2 = data[["team_name", "average_possession"]]

# Comparaison de la possession de balle moyenne à domicile et à l'extérieur pour chaque équipe
possession_data2.set_index("team_name", inplace=True)
possession_data2.plot(kind="bar", stacked=True, figsize=(12, 6))
plt.title("Comparaison de la possession de balle moyenne à domicile et à l'extérieur pour chaque équipe")
plt.xlabel("Équipe")
plt.ylabel("Possession de balle moyenne (%)")
plt.xticks(rotation=80)
plt.legend(title="Lieu du match", loc="upper right")
plt.tight_layout()
plt.show()



```
